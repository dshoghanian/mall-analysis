```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

customer_df = pd.read_excel('customer_data.xlsx')
sales_df = pd.read_excel('sales_data.xlsx')
mall_df = pd.read_excel('shopping_mall_data.xlsx')
```

To begin with, we loaded in our three datasets and looked at the first five rows in order to get a better idea of what the data could tell us. We pinpointed shared columns that we could merge on and any interesting trends we could find at first glance. Noticing that our customer and sales dataframes shared customer_id columns and our sales and malls dataframes shared shopping_mall columns, we decided to merge them together.

```python
cust_sales_df = pd.merge(customer_df,sales_df, on='customer_id')
cust_sales_df.head()
cust_sales_malls_df = pd.merge(cust_sales_df, mall_df, on='shopping_mall')
cust_sales_malls_df.head(20)
cust_sales_malls_df.set_index('customer_id', inplace=True)
cust_sales_malls_df['sale'] = cust_sales_malls_df['price'] * cust_sales_malls_df['quantity']
cust_sales_malls_df['sale']
cust_sales_malls_df.head()
cust_sales_malls_df.index.nunique()
cust_sales_malls_df.info()
```

After merging the data we noticed that there was a price and a quantity column, leading us to the realization that we needed a total sale revenue column in order to see how much was purchased in total during a single trip. Another observation we had at this point was the fact that there were no repeat customer_id's. This meant that customer records were not kept, just sales records from each purchase. This altered our original thought process of analyzing individual customers to analyzing customer profiles. After merging the data and getting an understanding of how it was structured, we decided to look at our first question.

What are the overall trends for malls in California
</div>

The overall trend we decided to focus on was sales over the course of the last three years. Our intital hypothesis was that we would see a downward trend in sales over the course of the three years that the datasets covered. In order to more clearly visualize this, we needed to add month and year columns to our dataset.

```python
cust_sales_malls_df['invoice date'] = pd.to_datetime(cust_sales_malls_df['invoice date'])
cust_month = cust_sales_malls_df.loc[cust_sales_malls_df['invoice date'].dt.month == 1,:]
months = {1:"January",2:"February",3:"March",4:"April",5:"May",6:"June",7:"July",
          8:"August",9:"September",10:"October",11:"November",12:"December"}
month = []
for i in range(1,13):
  month.append(months[i])
month
cust_sales_malls_df['month_of_sale'] = cust_sales_malls_df['invoice date'].dt.month.map(months)
```

```python
years = {2021: 21, 2022: 22, 2023: 23}
cust_sales_malls_df['year'] = cust_sales_malls_df['invoice date'].dt.year.map(years)
print("2023 Customer Frequency\n")
print(cust_sales_malls_df.loc[cust_sales_malls_df['year'] == 23,:]['shopping_mall'].value_counts())
print("\n2022 Customer Frequency\n")
print(cust_sales_malls_df.loc[cust_sales_malls_df['year'] == 22,:]['shopping_mall'].value_counts())
print("\n2021 Customer Frequency\n")
print(cust_sales_malls_df.loc[cust_sales_malls_df['year'] == 21,:]['shopping_mall'].value_counts())
```

When adding our year column, we decided to take a look at shopper frequency across the years. We noticed a strange pattern where shopper frequency looked regular in 2021 and 2022 but had a steep drop off in 2023. In order to look at this on a mall by mall basis we seperated each mall into it's own dataframe in a dictionary and looked at different columns to compare between the years.

```python
mall_dataframes = {}
cust_sales_malls_df[cust_sales_malls_df.year == 22]['shopping_mall'].value_counts()
cust_sales_malls_df[cust_sales_malls_df.year == 23]['shopping_mall'].value_counts()
for i in cust_sales_malls_df['shopping_mall'].unique():
  mall_dataframes[i] = cust_sales_malls_df[cust_sales_malls_df['shopping_mall'] == i]
plt.plot(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 23].groupby('month_of_sale')['sale'].sum(), label="2023")
plt.plot(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 22].groupby('month_of_sale')['sale'].sum(), label="2022")
plt.xticks(rotation=45)
plt.title('Total Revenue for Beverley Center 2022 vs. 2023')
plt.ylabel('Total Revenue')
plt.legend()
print(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 23]['category'].value_counts())
print(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 22]['category'].value_counts())
print(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 22]['payment_method'].value_counts())
print(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 23]['payment_method'].value_counts())
print(mall_dataframes['Beverly Center'][cust_sales_malls_df.year == 22]['shopping_mall'].value_counts())
```

After looking at frequency's and visualizing sales for the Beverley Center, we could see that this trend was present across the data. In order to get a better idea of the severity of the drop off, we decided to plot the data across the three years on the same graph.

```python
abvs = {"January":"Jan","February":"Feb","March":"Mar","April":"Apr","May":"May","June":"Jun",
        "July":"Jul","August":"Aug","September":"Sep","October":"Oct","November":"Nov","December":"Dec"}
ordered_monyear = ['Jan 21', 'Feb 21', 'Mar 21', 'Apr 21', 'May 21', 'Jun 21', 'Jul 21', 'Aug 21', 'Sep 21', 'Oct 21', 'Nov 21', 'Dec 21',
           'Jan 22', 'Feb 22', 'Mar 22', 'Apr 22', 'May 22', 'Jun 22', 'Jul 22', 'Aug 22', 'Sep 22', 'Oct 22', 'Nov 22', 'Dec 22',
           'Jan 23', 'Feb 23', 'Mar 23', 'Apr 23', 'May 23', 'Jun 23', 'Jul 23', 'Aug 23', 'Sep 23', 'Oct 23', 'Nov 23', 'Dec 23']
cust_sales_malls_df['abvs'] = cust_sales_malls_df['month_of_sale'].map(abvs)
cust_sales_malls_df['monthyear'] = cust_sales_malls_df['abvs'] + " " + cust_sales_malls_df['year'].astype(str)
timeseries = cust_sales_malls_df.groupby('monthyear')['sale'].sum().reset_index()
timeseries['monthyear'] = pd.Categorical(timeseries['monthyear'], categories=ordered_monyear, ordered=True)
timeseries.sort_values(by='monthyear', inplace=True)
fig, ax3 = plt.subplots()
ax3.plot(timeseries['monthyear'], timeseries['sale'])
ax3.set_xticklabels(labels=timeseries['monthyear'],rotation=90)
ax3.set_ylabel('Total Revenue (in tens of millions)')
ax3.set_title('Total Revenue for California Malls 2021 - 2023')
```

This graph shows how drastic the drop off in sales was from 2022 to 2023. At this point we weren't sure if this was simply due to a lack of data in the dataset or if this was a real trend in the shopping mall industry. To ensure this trend wasn't occuring due to a just a few malls, we decided to plot each of the malls across the three years.

```python
fig, ax4 = plt.subplots(5,2)
malls = list(cust_sales_malls_df['shopping_mall'].unique())
for i in cust_sales_malls_df['shopping_mall'].unique():
  mall_dataframes[i] = cust_sales_malls_df[cust_sales_malls_df['shopping_mall'] == i]
for mall in malls:
  df = mall_dataframes[mall]
  name = df.iloc[0,8]
  df = df.groupby('monthyear')['sale'].sum().reset_index()
  df['monthyear'] = pd.Categorical(df['monthyear'], categories=ordered_monyear, ordered=True)
  df.sort_values(by='monthyear', inplace=True)
  plt.ylabel('Total Revenue (in tens of millions)')
  plt.plot(df['monthyear'],df['sale'],label=name)
  plt.legend(prop={'size':6})
plt.xticks(ticks=ordered_monyear,rotation=90)
plt.title('Total Revenue by Mall in California 2021 - 2023')
```

Here we see that this trend is present across all shopping malls in the dataset. In our interpretations section we will cover the external research we did in order to cross refrence this trend and how we decided to understand our findings. After our initial exploratory analysis, we decided to dive deeper into our sales data in order to uncover the trends in shopper behavior.

<div style="text-align: center; font-size 12px; font-weight: bold;">
What seasonality/cyclicality is present in shopper behavior
</div>

```python
monthly_sales = cust_sales_malls_df.groupby('month_of_sale')['sale'].sum().reset_index()
fig, ax = plt.subplots(2,1)
monthly_sales['month_of_sale'] = pd.Categorical(monthly_sales['month_of_sale'],categories=month,ordered=True)
monthly_sales.sort_values(by='month_of_sale', inplace=True)
ax[1].bar(monthly_sales['month_of_sale'],monthly_sales['sale'])
ax[0].plot(monthly_sales['month_of_sale'],monthly_sales['sale'])
ax[0].set_xticklabels(labels=[])
ax[0].set_ylabel('Revenue (in tens of millions)')
ax[1].set_xticklabels(labels=monthly_sales['month_of_sale'],rotation=45)
ax[1].set_ylabel('Revenue (in tens of millions)')
ax[0].set_title('Revenue by Month')
```

The revenue trends show a decline after January, possibly due to post holiday reduced spending. The pattern reflects peak in months of November and December (holiday season). Reduced shopping in months January - March as people recover from holiday spending

```python
seasons = {"January":"Winter","February":"Winter","March":"Spring","April":"Spring","May":"Spring","June":"Summer","July":"Summer",
          "August":"Summer","September":"Fall","October":"Fall","November":"Fall","December":"Winter"}
cust_sales_malls_df['seasons'] = cust_sales_malls_df['month_of_sale'].map(seasons)
seasons_sorted = ['Winter','Spring','Summer','Fall']
seasonal_sales = cust_sales_malls_df.groupby('seasons')['sale'].sum().reset_index()
seasonal_sales
seasonal_sales['seasons'] = pd.Categorical(seasonal_sales['seasons'],categories=seasons_sorted,ordered=True)
seasonal_sales.sort_values(by='seasons',inplace=True)
fig, ax2 = plt.subplots(2,1)
ax2[0].plot(seasonal_sales['seasons'],seasonal_sales['sale'])
ax2[1].bar(seasonal_sales['seasons'],seasonal_sales['sale'])
ax2[0].set_ylabel('Revenue (in tens of millions)')
ax2[1].set_ylabel('Revenue (in tens of millions)')
```

Winter (December - February) shows the highest activity, aligning with the holiday season. A gradual decline in spring and summer suggests reduced shopping intensity. Fall shows a slight increase as the holiday season begins.

```python

sales_df['invoice date'] = pd.to_datetime(sales_df['invoice date'])
sales_df['day_of_week'] = sales_df['invoice date'].dt.day_name()
shoppers_by_day = sales_df.groupby('day_of_week').size().reset_index(name='shopper_count')
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
shoppers_by_day['day_of_week'] = pd.Categorical(shoppers_by_day['day_of_week'], categories=days_order, ordered=True)
shoppers_by_day.sort_values('day_of_week', inplace=True)



plt.bar(shoppers_by_day['day_of_week'], shoppers_by_day['shopper_count'], color='skyblue')
plt.title('Shoppers Count by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Shoppers')
plt.xticks(rotation=45)
plt.show()
```

Shopping activity is generally consistent across weekdays (Monday - Friday). Weekends (Saturday and Sunday) see similar or slightly lower shopper counts compared to the weekdays. Possibly due to leisure-oriented shopping trips.

```python
all_malls_data = sales_df.groupby(['shopping_mall', 'day_of_week']).size().reset_index(name='shopper_count')
all_malls_data['day_of_week'] = pd.Categorical(all_malls_data['day_of_week'], categories=days_order, ordered=True)
pivot_data = all_malls_data.pivot(index='day_of_week', columns='shopping_mall', values='shopper_count').fillna(0)
pivot_data.plot(kind='bar', figsize=(12, 7), width=0.8)
plt.title('Shoppers Count by Day of the Week for All Malls')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Shoppers')
plt.xticks(rotation=45)
plt.legend(title='Shopping Mall', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Malls like South Coast Plaza and Beverly Center show consistent shopper counts throughout the week, with no clear weekend spike. Smaller malls like The Grove or Fashion Valley might have slightly reduced weekday activity

```python
mall_summary = cust_sales_malls_df.groupby('shopping_mall').agg(
    total_revenue=('sale', 'sum'),
    avg_purchase_price=('sale', 'mean'),
    store_count=('store_count', 'mean'),
    mall_size=('area (sqm)', 'mean'),
    mall_age=('construction_year', lambda x: 2024 - x.mean())
).reset_index()
```

```python
plt.scatter(mall_summary['mall_size'], mall_summary['total_revenue'], color='b', alpha=0.6)
plt.title('Mall Size vs. Total Revenue')
plt.xlabel('Mall Size (sqm)')
plt.ylabel('Total Revenue')
plt.show()
```

In this chart we can see there is a trend of larger Total Revenue correlated with larger Mall size.

```python
plt.scatter(mall_summary['store_count'], mall_summary['total_revenue'], color='b', alpha=0.6)
plt.title('Store Count vs. Total Revenue')
plt.xlabel('Store Count')
plt.ylabel('Total Revenue')
plt.show()
```

In this chart we can see there is a trend of larger Total Revenue correlated with larger Store Counts. 

```python
plt.scatter(mall_summary['mall_age'], mall_summary['total_revenue'], color='b', alpha=0.6)
plt.title('Mall Age vs. Total Revenue')
plt.xlabel('Mall Age (Years)')
plt.ylabel('Total Revenue')
plt.show()
```

In this chart we can see there is a trend of larger Total Revenue correlated with older malls. 

```python
plt.scatter(mall_summary['mall_age'], mall_summary['mall_size'], label='Mall Size', color='blue', alpha=0.6)
plt.xlabel('Mall Age (Years)')
plt.ylabel('Mall Size (sqm)')
plt.title('Mall Age vs. Mall Size')
plt.show()
```

We can see in this chart that older malls tend to also be larger in size.

```python
category_mall_stats = cust_sales_malls_df.groupby(['shopping_mall', 'category'])['sale'].sum().reset_index()
pivot_data = category_mall_stats.pivot(index='shopping_mall', columns='category', values='sale')
category_medians = cust_sales_malls_df.groupby('category')['sale'].median()
pivot_data = pivot_data.apply(lambda col: col.fillna(category_medians[col.name]), axis=0)
categories = cust_sales_malls_df['category'].unique()
category_colors = sns.color_palette('tab10', len(categories))
pivot_data.plot(kind='bar',stacked=True,figsize=(12, 6),color=category_colors, alpha=0.85)
plt.title('Total Revenue by Category Across Malls', fontsize=16)
plt.xlabel('Shopping Mall')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category', loc='upper left')
plt.show()
```

This chart shows that the percentage of Total Revenue by Category is consistent across malls. Clothing, Shoes, and Technology comprise the majority of Total Revenue.

```python
cust_sales_malls_df['age'].describe()
bins=[18,28,38,48,58,69]
labels=['18-28','29-38','39-48','49-58','59-69']
cust_sales_malls_df['age_group'] = pd.cut(cust_sales_malls_df['age'],bins=bins,labels=labels,right=True,include_lowest=True)
gender_age_mall = cust_sales_malls_df.groupby(['shopping_mall', 'gender', 'age_group']).size().reset_index(name='count')
gender_age_mall = gender_age_mall.dropna(subset=['gender', 'age_group'])
pivot_data = gender_age_mall.pivot_table(index='shopping_mall', columns=['gender', 'age_group'], values='count')
malls = pivot_data.index
age_ranges = cust_sales_malls_df['age_group'].unique()
bar_width = 0.4
x = np.arange(len(malls))
gender_colors = {'Male': 'lightblue', 'Female': 'lightcoral'}
fig, ax = plt.subplots(figsize=(12, 8))
genders = ['Male','Female']
for i, gender in enumerate(genders):
    bottom = np.zeros(len(malls)) 
    for age_group in age_ranges:
        age_counts = pivot_data[(gender, age_group)].values if (gender, age_group) in pivot_data.columns else np.zeros(len(malls))
        bars = ax.bar(x + (i * bar_width), age_counts, bar_width, color=gender_colors[gender], edgecolor='white', bottom=bottom)
        for bar, count in zip(bars, age_counts):
            if count > 0:
                ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2,str(age_group), ha='center', va='center', fontsize=8, color='black')
        bottom += age_counts
ax.set_xticks(x + bar_width / 2)
ax.set_xlabel('Shopping Mall')
ax.set_ylabel('Shoppers')
ax.set_xticklabels(malls, rotation=45, ha='right')
ax.set_title('Gender and Age Range Distribution Across Malls')
ax.legend([plt.Rectangle((0, 0), 1, 1, color=gender_colors[g]) for g in genders], genders, title='Gender', loc='upper left')
plt.show()
```

We can see in this chart that the breakdown of gender and age of shoppers is very consistent across malls.
